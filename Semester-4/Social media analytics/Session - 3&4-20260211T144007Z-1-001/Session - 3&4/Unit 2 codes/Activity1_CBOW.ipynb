```python
from gensim.models import Word2Vec
import random

sentences = [["i","love","nlp"],["nlp","is","cool"],["i","love","ai"]]

# Normal CBOW
model1 = Word2Vec(sentences, sg=0, vector_size=50, min_count=1, epochs=50)

# Corrupt ONLY first sentence so vocabulary remains
corrupt = sentences.copy()

s = corrupt[0].copy()
choices = [w for w in s if w != s[1]]
s[1] = random.choice(choices)
corrupt[0] = s

# Train corrupted model
model2 = Word2Vec(corrupt, sg=0, vector_size=50, min_count=1, epochs=50)

print("Original:", sentences)
print("Corrupted:", corrupt)

print("\nSimilarity before:", model1.wv.similarity("i","love"))
print("Similarity after:", model2.wv.similarity("i","love"))
```
