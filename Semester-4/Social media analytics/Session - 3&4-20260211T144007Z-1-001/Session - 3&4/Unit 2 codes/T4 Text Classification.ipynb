```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

# Small topic dataset
texts = [
    # Technology
    "AI is transforming the world",
    "Machine learning models are powerful",
    "Deep learning improves computer vision",

    # Politics
    "Government announced new policy",
    "Elections will be held next month",
    "Parliament passed the bill",

    # Sports
    "Football match was exciting",
    "Cricket team won the series",
    "Player scored a great goal"
]

# Labels: 0=Tech, 1=Politics, 2=Sports
labels = [0,0,0, 1,1,1, 2,2,2]

label_map = {
    0: "Technology",
    1: "Politics",
    2: "Sports"
}

df = pd.DataFrame({"text":texts, "label":labels})

# TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["text"])

# Train Naive Bayes on ALL data (demo purpose)
model = MultinomialNB()
model.fit(X, labels)

# New sentences to classify
test_sentences = [
    "New AI model released",
    "Prime minister announced election",
    "Team won the football match"
]

X_test = vectorizer.transform(test_sentences)
pred = model.predict(X_test)

# Print readable results
for s, p in zip(test_sentences, pred):
    print(s, "→", label_map[p])
```
