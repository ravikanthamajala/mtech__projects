```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love NLP",
    "I love AI",
    "NLP is interesting"
]

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(documents)

print("Vocabulary:")
print(tfidf.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(X.toarray())
```
