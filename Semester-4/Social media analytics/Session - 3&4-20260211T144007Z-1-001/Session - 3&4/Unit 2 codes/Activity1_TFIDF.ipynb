```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

corpus = [
    "I love NLP",
    "NLP is interesting",
    "I love AI",
    "AI is powerful"
]

tfidf = TfidfVectorizer()
X = tfidf.fit_transform(corpus)

idf = tfidf.idf_
reverse_score = 1 / idf

words = tfidf.get_feature_names_out()

least = np.argsort(reverse_score)[-5:]

print("Least informative words:")
print([words[i] for i in least])
```
