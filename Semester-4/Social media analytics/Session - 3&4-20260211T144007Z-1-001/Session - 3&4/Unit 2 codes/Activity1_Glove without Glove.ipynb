```python
import numpy as np
from sklearn.decomposition import TruncatedSVD

sent = [["i","love","nlp"],["nlp","is","fun"],["i","love","ai"]]
vocab = list(set(w for s in sent for w in s))
idx = {w:i for i,w in enumerate(vocab)}

mat = np.zeros((len(vocab),len(vocab)))

for s in sent:
    for i,w in enumerate(s):
        for j in range(len(s)):
            if i!=j:
                mat[idx[w]][idx[s[j]]] +=1

svd = TruncatedSVD(3)
emb = svd.fit_transform(mat)

def sim(w):
    v=emb[idx[w]]
    sims=[]
    for k in vocab:
        sims.append((k,np.dot(v,emb[idx[k]])))
    return sorted(sims,key=lambda x:-x[1])[1:4]

print(sim("nlp"))
```
