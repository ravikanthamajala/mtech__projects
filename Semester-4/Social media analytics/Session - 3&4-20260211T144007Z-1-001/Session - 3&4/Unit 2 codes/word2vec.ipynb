```python
!pip install gensim

from gensim.models import Word2Vec

# Sample corpus
sentences = [
    ["i", "love", "nlp"],
    ["i", "love", "ai"],
    ["nlp", "is", "interesting"],
    ["ai", "is", "powerful"]
]

# Train CBOW Model (sg = 0)
cbow_model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4,
    sg=0
)

# Train Skip-Gram Model (sg = 1)
skipgram_model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4,
    sg=1
)

# Get vectors
cbow_vector = cbow_model.wv["nlp"]
skipgram_vector = skipgram_model.wv["nlp"]

print("CBOW Vector for 'nlp':")
print(cbow_vector)

print("\nSkip-Gram Vector for 'nlp':")
print(skipgram_vector)
```
