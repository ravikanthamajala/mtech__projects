```python
!pip install gensim

from gensim.models import Word2Vec
import numpy as np

sent = [["king","queen","man","woman","royal"]]

model = Word2Vec(sent, vector_size=50, min_count=1, epochs=100)

# Simulate missing vector by zeroing it
model.wv.vectors[model.wv.key_to_index["queen"]] = np.zeros(model.vector_size)

print("Similar words to 'king' after removing 'queen':")
print(model.wv.most_similar("king"))
```
