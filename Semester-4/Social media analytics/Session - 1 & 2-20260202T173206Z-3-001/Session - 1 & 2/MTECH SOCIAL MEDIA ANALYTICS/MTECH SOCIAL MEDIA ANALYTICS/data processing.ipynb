```python
# 1) Tokenization
from nltk.tokenize import word_tokenize

text = "The camera quality of this phone is amazing!"
tokens = word_tokenize(text)

print("Input:", text)
print("Tokens:", tokens)
```

```python
# 2) Stop word removal
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

filtered = [w for w in tokens if w.lower() not in stop_words]

print("After Stopword Removal:", filtered)
```

```python
# 3) Stemming
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["playing", "played", "plays", "player"]
stemmed = [stemmer.stem(w) for w in words]

print("Input:", words)
print("Stemmed:", stemmed)
```

```python
# 4) Lemmatization
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

words = ["better", "running", "phones"]
lemmatized = [lemmatizer.lemmatize(w) for w in words]

print("Input:", words)
print("Lemmatized:", lemmatized)
```

```python
# 5) Text normalization
import re

text = "OMG!!! This phone's camera is gr8 😍"

normalized = text.lower()

# Step 1: Expand slang FIRST
normalized = normalized.replace("gr8", "great")

# Step 2: Remove emojis & special characters
normalized = re.sub(r"[^a-zA-Z\s]", "", normalized)

print("Original:", text)
print("Normalized:", normalized)
```

```python
# 6) Emoji handling
import emoji

text = "Camera quality is amazing 😍🔥"
converted = emoji.demojize(text)

print("Original:", text)
print("With Emojis Converted:", converted)
```

```python
# 7) Sentiment analysis
review = "The battery life is terrible but the camera is excellent"
print("Review:", review)
```

```python
# (A) TextBlob – Polarity & Subjectivity
from textblob import TextBlob

blob = TextBlob(review)

print("Polarity:", blob.sentiment.polarity)
print("Subjectivity:", blob.sentiment.subjectivity)
```

```python
# (B) VADER – Positive/Negative/Neutral
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()
scores = sia.polarity_scores(review)

print(scores)
```

```python
# (C) Final Label
if scores["compound"] > 0:
    print("Sentiment: Positive")
elif scores["compound"] < 0:
    print("Sentiment: Negative")
else:
    print("Sentiment: Neutral")
```

```python
# 8) Feature engineering
# Input
reviews = [
    "Camera is amazing",
    "Battery is terrible",
    "Display quality is good"
]
```

```python
# (A) Bag of Words (BoW)
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(reviews)

print("Features:", vectorizer.get_feature_names_out())
print("BoW Matrix:\n", bow.toarray())
```

```python
# (B) TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(reviews)

print(tfidf_matrix.toarray())
```

```python
# (C) N-grams (Bigrams + Trigrams)
ngram = CountVectorizer(ngram_range=(2,3))
ngram_matrix = ngram.fit_transform(reviews)

print("N-gram Features:", ngram.get_feature_names_out())
```

```python
# 9. AI EMBEDDINGS
# (A) Word2Vec
from gensim.models import Word2Vec

reviews = [
    "Camera is amazing",
    "Battery is terrible",
    "Display quality is good"
]

# Convert to lowercase + tokenize
sentences = [r.lower().split() for r in reviews]

model = Word2Vec(sentences, vector_size=50, min_count=1)

print("Vocabulary:", model.wv.index_to_key)
print("\nEmbedding for 'camera':")
print(model.wv["camera"])
```

```python
# (B) BERT Embeddings (Contextual)
from sentence_transformers import SentenceTransformer

bert = SentenceTransformer("all-MiniLM-L6-v2")
embedding = bert.encode("Camera is amazing")

print("Embedding shape:", embedding.shape)
```
