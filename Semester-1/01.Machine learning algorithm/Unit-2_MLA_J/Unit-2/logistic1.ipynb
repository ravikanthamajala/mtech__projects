**Logistic Regression Implementation**

```python
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Create Dataset**

```python
from sklearn.datasets import make_classification
```

```python
#create the datasets
x,y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=42)
```

```python
x.shape
```

```python
y.shape
```

```python
pd.DataFrame(x)
```

```python
y
```

```python
#Train test Data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#Model Training
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
#prediction
y_pred=model.predict(x_test)
```

```python
print(y_pred)
```

**Perfomance Metrics**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test,y_pred)  # accuracy
print(score)
cm=confusion_matrix(y_test,y_pred)     # Confusion Matrix
print(cm)
cr=classification_report(y_test,y_pred)  #Precision,Recall,F1-Score
print(cr)
```

# Hyper parameter tuning and Cross Valiadation

# Gridsearch CV

```python
#Hyper parameters
model=LogisticRegression()
penalty_var=('l1','l2','elasticnet','none')
solver_var=('newton-cg','lbfgs','liblinear','sag','saga')
c_values=[100,10,1.0,0.1,0.01]
```

```python
params = dict(penalty=penalty_var,C=c_values,solver=solver_var)
```

```python
from sklearn.model_selection import StratifiedKFold
cv_var=StratifiedKFold(n_splits=5,shuffle=False,random_state=None)
```

```python
#GridSearch CV
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv_var,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(x_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_test,y_pred)  # accuracy
print(score)
cm=confusion_matrix(y_test,y_pred)     # Confusion Matrix
print(cm)
cr=classification_report(y_test,y_pred)  #Precision,Recall,F1-Score
print(cr)
```

# Randomized search cv

```python
from sklearn.model_selection import RandomizedSearchCV
model=LogisticRegression()
random=RandomizedSearchCV(estimator=model,param_distributions=params,scoring='accuracy',cv=5,n_jobs=-1)
```

```python
random.fit(x_train,y_train)
```

```python
random.best_params_
```

```python
random.best_score_
```

```python
y_pred=random.predict(x_test)
```
