Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Sample data (x=years of experience , y= salary(1000$))
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
```

```python
# Reshape x to be a 2D array as required by scikit-learn
x = np.array(x).reshape(-1, 1)  # Reshape x to a column vector
y = np.array(y)
```

```python
print(x)
```

```python
print(y)
```

```python
# Create and fit the linear regression model
model = LinearRegression()
model.fit(x, y)
```

```python
# Predict y values
y_pred = model.predict(x)
```

```python
print(y_pred)
```

```python
# Coefficients
print("Slope (Coefficient):", model.coef_)
print("Intercept:", model.intercept_)
```

```python
# Plotting the results
plt.scatter(x, y, color='red',label='Actual data')  # Original data points
plt.plot(x, y_pred, color='blue',label='Fitted line')  # Regression line
plt.scatter(x, y_pred, color ='green',marker='o',label='predicted dat')
#Annotate predicted points with coordinates
for i in range(len(x)):
    plt.annotate(f'({x[i][0]},{y_pred[i]})', (x[i][0], y_pred[i]), textcoords="offset points", xytext=(0,10), ha='center')
plt.title('x vs y (Linear Regression)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

**Mean Absolute Error**

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
# Calculate MAE using sklearn
mae = mean_absolute_error(y,y_pred)
mse = mean_squared_error(y,y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```

**Mean Squared Error (MSE):**

1.Compute the difference between each pair of actual and predicted values.

2.Square each difference.

3.Find the average of these squared differences.

```python
# Calculate the Mean Squared Error Manually
mse = np.mean((y - y_pred) ** 2)
print(f"Mean Squared Error: {mse}")
```

**Root Mean Square Error (RMSE)**

```python
# Calculate RMSE Manually
rmse = np.sqrt(mse)
print(f"RMSE: {rmse}")
```

For a target value scale of 5 , an RMSE of ***0.69*** is relatively low, suggesting that the model is performing reasonably well.
