```python
xbar = 24
mu = 25
n = 45
sigma =  2.5

zstat =  (xbar - mu)/(sigma/ np.sqrt(n))
zstat
```

```python
alpha = 0.01
alphaby2 = alpha/2
zalphaby2 = st.norm.isf(alphaby2)
zalphaby2 
```

```python
alphaby2
```

```python
std_error =  sigma / np.sqrt(n)
ME = zalphaby2 * std_error
ME
```

```python
lci = xbar - ME
uci = xbar + ME
print (lci,uci)
```

```python
st.norm.interval(0.99,xbar,std_error)
```
