```python
data = {'Values': [50, 20, 30, 40, 10, 20],}
df = pd.DataFrame(data)
```

```python
df['Rank_Average'] = df['Values'].rank(method='average',ascending=False)
df['Rank_Min'] = df['Values'].rank(method='min',ascending=False)
df['Rank_Max'] = df['Values'].rank(method='max',ascending=False)
df['Rank_First'] = df['Values'].rank(method='first',ascending=False)
df['Rank_Dense'] = df['Values'].rank(method='dense',ascending=False)

print(df)
```
