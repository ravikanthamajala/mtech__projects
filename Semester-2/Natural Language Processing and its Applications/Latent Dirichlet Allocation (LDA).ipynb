```python
import gensim
import gensim.corpora as corpora
import nltk
import spacy
from nltk.corpus import stopwords
from gensim.models import LdaModel
from gensim.utils import simple_preprocess
from pprint import pprint
```

```python
# Download stopwords if not available
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
# Sample Documents
documents = [
    "I like to eat broccoli and bananas.",
    "I ate a banana and salad for breakfast.",
    "Puppies and kittens are cute.",
    "My sister adopted a kitten yesterday.",
    "Look at this cute hamster munching on a piece of broccoli."
]
```

```python
# Preprocessing function
def preprocess_text(texts):
    """Tokenizes, removes stopwords, and lemmatizes the input text."""
    nlp = spacy.load("en_core_web_sm")
    processed_texts = []

    for text in texts:
        doc = nlp(text.lower())  # Convert to lowercase
        tokens = [token.lemma_ for token in doc if token.text not in stop_words and token.is_alpha]
        processed_texts.append(tokens)

    return processed_texts
```

```python
# Preprocess documents
processed_docs = preprocess_text(documents)

# Create dictionary and corpus
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
# Train LDA model
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=2, random_state=42, passes=10)
```

```python
# Print Topics
pprint(lda_model.print_topics())
```

```python
# Assign topics to each document
for i, doc in enumerate(corpus):
    print(f"Document {i+1} Topics:", lda_model.get_document_topics(doc))
```
