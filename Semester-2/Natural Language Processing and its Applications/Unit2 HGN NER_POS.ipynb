```python
# Load pre-trained model (e.g., 'en_core_web_sm')
nlp=spacy.load("en_core_web_sm")
```

## 2. NER

```python
# Example text
text="Mr.Raju is an American politician who served as the 44th President of the United States from 2009 to 2017. He is the first African American to have served as president, as well as the first born outside the contiguous United States."
```

```python
# Process the text
doc3=nlp(text)
```

What Happens Internally?
Tokenization: Splits the text into words, punctuation, etc.

POS Tagging: Assigns grammatical categories (noun, verb, etc.).

Dependency Parsing: Identifies relationships between words.

Named Entity Recognition (NER): Identifies proper names, organizations, dates, etc.

Lemmatization: Converts words to their base forms.


```python
from spacy import displacy
displacy.render(doc3, style='ent', jupyter=True)
```

```python
# Extract entities
entities = [(ent.text, ent.label_) for ent in doc3.ents]
print(entities)
```

```python
import pandas as pd
# Example: Creating a DataFrame
data = []
for ent in doc3.ents:
    data.append({"Entity": ent.text, "Label": ent.label_})
df = pd.DataFrame(data)
print(df)
```

## 3. Part of Speech (POS) Tagging

```python
#Process a sentence
doc = nlp('She saw a bear. She was very afraid.')
```

```python
type(doc)
```

```python
#Print POS tags for each word 2 for word in doc:
for word in doc:
  print (word, word.pos_)
```

```python
#Check another document
doc2 = nlp('Your effort will bear fruit.')
for word in doc2:
  print (word, word.pos_)
```

## 4. Dependency Parsing in spaCy

```python
# Example sentence
text = "The quick brown fox jumps over the lazy dog."
# Process text
doc = nlp(text)
# Print dependencies
for token in doc:
    print(f"{token.text} → {token.dep_} → {token.head.text}")
```

```python
from spacy import displacy
# Render the dependency tree
displacy.render(doc, style="dep", jupyter=True)
```
