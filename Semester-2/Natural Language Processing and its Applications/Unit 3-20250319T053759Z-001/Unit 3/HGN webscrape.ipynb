```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
# URL of the webpage containing the table
url = "https://www.worldometers.info/coronavirus/"

# Send a request to fetch the webpage content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print(f"Failed to retrieve webpage, status code: {response.status_code}")
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table with the id 'main_table_countries_today'
table = soup.find("table", id="main_table_countries_today")

# Extract the table rows
rows = table.find_all("tr")

# Create empty lists to store headers and data
headers = []
data = []

# Extract table headers
for th in rows[0].find_all("th"):
    headers.append(th.text.strip())

# Extract table data row by row
for row in rows[1:]:
    row_data = [td.text.strip() for td in row.find_all("td")]
    if row_data:
        data.append(row_data)

# Create a Pandas DataFrame to store the scraped data
df = pd.DataFrame(data, columns=headers)

# Show the first 5 rows of the DataFrame
print(df.head())
print(df.info())
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage containing the table
url = "https://www.worldometers.info/coronavirus/"

# Send a request to fetch the webpage content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the webpage!")

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Extract headline text (e.g., h1 and h2 tags)
    print("Headlines on the page:")
    for heading in soup.find_all(['h1', 'h2']):
        print(heading.text.strip())

    # Extract image URLs
    print("\nImages on the page:")
    images = soup.find_all("img")
    for img in images:
        img_url = img.get("src")
        if img_url:
            print(f"Image URL: {img_url}")

else:
    print(f"Failed to retrieve webpage, status code: {response.status_code}")
```

```python
import requests

# URL for fetching Bitcoin price from CoinGecko API
api_url = "https://api.coingecko.com/api/v3/simple/price?ids=bitcoin&vs_currencies=usd"

# Send a GET request to the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    bitcoin_data = response.json()  # Parse JSON response
    print("Bitcoin Price in USD:", bitcoin_data['bitcoin']['usd'])
else:
    print(f"Failed to retrieve data, status code: {response.status_code}")
```

```python
import requests
import pandas as pd

# World Bank API endpoint to get GDP data for all countries for 2021
api_url = "http://api.worldbank.org/v2/country/all/indicator/NY.GDP.MKTP.CD?date=2021&format=json"

# Send a GET request to fetch the data
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the JSON response
    data = response.json()[1]  # The data we want is in the second element of the response
   # data = response.json()
    #print(data)
    # Extract relevant information (Country, GDP)
    countries = []
    gdps = []

    for entry in data:
        country_name = entry["country"]["value"]
        gdp = entry["value"]

        countries.append(country_name)
        gdps.append(gdp)

    # Create a DataFrame to display the data in a tabular format
    df = pd.DataFrame({"Country": countries, "GDP (USD)": gdps})
    print(df.head())  # Display first 5 rows of the DataFrame
```

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Specify the URL of the webpage to scrape
url = "https://en.wikipedia.org/wiki/software testing"

# Step 2: Send an HTTP request to the webpage
response = requests.get(url)

# Step 3: Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the webpage!")

    # Step 4: Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    # Step 5: Extract text from <p> tags
    paragraphs = soup.find_all('p')
    for para in paragraphs:
        print(para.get_text())
else:
    print(f"Failed to fetch the webpage. Status code: {response.status_code}")
```
