## 2. Create Training & Test Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
# split X and y into training and testing sets
sms_train, sms_test, y_train, y_test = train_test_split(sms_df.message,
                                                        sms_df.label_num,
                                                        random_state=2)
```

```python
sms_train.reset_index(inplace=True, drop=True)
sms_test.reset_index(inplace=True, drop=True)
y_train.reset_index(inplace=True, drop=True)
y_test.reset_index(inplace=True, drop=True)
```

```python
#Traing data
print(sms_train.shape)
print(y_train.shape)
```

```python
#Test Data
print(sms_test.shape)
print(y_test.shape)
```

## 3. Tokenization & Vectorization

Using **CountVectorizer**, to get numeric features.

```python
# import and instantiate CountVectorizer (with the default parameters)
from sklearn.feature_extraction.text import CountVectorizer
cvect = CountVectorizer()
```

```python
sms_train.head()
```

```python
#Feed SMS data to CountVectorizer
cvect.fit(sms_train)
```

```python
#Check the vocablury size
len(cvect.vocabulary_)
```

```python
#What is there in the vocabulary
cvect.vocabulary_
```

Build Document-term Matrix (DTM)

```python
#Convert Training SMS messages into Count Vectors
X_train_ct = cvect.transform(sms_train)
```

```python
#Size of Document Term Matrix
X_train_ct.shape
```

```python
#Let's check the first record
sms_train[0]
```

```python
# let's check first record
X_train_ct[0]
```

```python
print(X_train_ct[0:1])
```

```python
X_test_ct = cvect.transform(sms_test)
X_test_ct.shape
```


### Summary:
1)vect.fit(train) learns the **vocabulary** of the training data.

2) vect.transform(train) uses the **fitted vocabulary** to build a **document-term matrix** from the training data.

3) vect.transform(test) uses the **fitted vocabulary** to build a document-term matrix from the testing data and **ignores tokens** it hasn't seen before.

### 8.Controlling Vocabulary Size

Thus far, we have been using the default parameter of CountVectorizer

```python
cvect.vocabulary_
```

```python
#?cvect
```

**max_df**: float in range [0.0, 1.0] or int, default=1.0

• When building the vocabulary, ignore terms that have a document frequency strictly higher than the given threshold (corpus-specific stop words).
o If float, the parameter represents a proportion of documents.
• If integer, the parameter represents an absolute count.

```python
# ignore terms that appear in more than 50% of the documents
vect = CountVectorizer(max_df=0.1)
vect.fit(sms_train)
len(vect.get_feature_names_out())
```

   min_df: int, default=1

```python
# only keep terms that apper in at least 2 documents
vect=CountVectorizer(min_df=2)
vect.fit(sms_train)
len(vect.get_feature_names_out())
```
