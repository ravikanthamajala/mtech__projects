```python
import tensorflow as tf
import numpy as np

# Define a simple dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)  # Inputs
y = np.array([[0], [1], [1], [0]], dtype=np.float32)  # XOR outputs

# Define a simple neural network
class NeuralNetwork(tf.keras.Model):
    def __init__(self):
        super(NeuralNetwork, self).__init__()
        self.hidden_layer = tf.keras.layers.Dense(4, activation='sigmoid')  # Hidden layer
        self.output_layer = tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer

    def call(self, inputs):
        x = self.hidden_layer(inputs)
        return self.output_layer(x)

# Initialize the model and optimizer
model = NeuralNetwork()
optimizer = tf.keras.optimizers.SGD(learning_rate=0.1)
loss_function = tf.keras.losses.BinaryCrossentropy()

# Training loop
epochs = 1000
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        # Forward pass
        predictions = model(X)
        # Compute the loss
        loss = loss_function(y, predictions)
    
    # Compute gradients (backpropagation)
    gradients = tape.gradient(loss, model.trainable_variables)
    # Update weights
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))

    # Print loss every 100 epochs
    if (epoch + 1) % 100 == 0:
        print(f"Epoch {epoch + 1}, Loss: {loss.numpy():.4f}")

# Test the model
predictions = model(X)
print("\nPredictions:")
print(tf.round(predictions))
```
