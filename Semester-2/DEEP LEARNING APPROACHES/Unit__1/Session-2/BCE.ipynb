```python
import numpy as np
import tensorflow as tf

# Define a simple dataset
X = np.array([[0], [1], [2], [3]], dtype=np.float32)  # Example input features
y = np.array([[0], [1], [1], [0]], dtype=np.float32)  # True labels

# Define a simple model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, activation='sigmoid')  # Output layer with sigmoid activation
])

# Compile the model with BCE loss
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=10, verbose=1)
```
