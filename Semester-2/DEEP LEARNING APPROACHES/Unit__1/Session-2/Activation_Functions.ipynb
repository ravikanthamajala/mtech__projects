```python
import numpy as np

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example input
x = np.array([-2, -1, 0, 1, 2])  # A range of inputs

# Apply the sigmoid function
sigmoid_output = sigmoid(x)

# Print the result
print("Input:", x)
print("Sigmoid Output:", sigmoid_output)
```

```python
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-10, 10, 100)  # Generate 100 points between -10 and 10

# Apply sigmoid function
y = sigmoid(x)

# Plot the sigmoid function
plt.plot(x, y, label="Sigmoid Function")
plt.title("Sigmoid Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Sigmoid Output (σ(x))")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np

# Define the softmax function
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max(x) for numerical stability
    return exp_x / np.sum(exp_x)

# Example input (logits for 3 classes)
logits = np.array([2.0, 1.0, 0.1])

# Apply the softmax function
softmax_output = softmax(logits)

# Print the results
print("Logits:", logits)
print("Softmax Output:", softmax_output)
print("Sum of Softmax Outputs:", np.sum(softmax_output))  # Should be 1
```

```python
import matplotlib.pyplot as plt

# Define a range of logits
logits_range = np.linspace(-2, 2, 100)
logits = np.vstack([logits_range, logits_range + 1, logits_range - 1]).T

# Apply softmax to each set of logits
softmax_outputs = np.apply_along_axis(softmax, 1, logits)

# Plot the softmax outputs
plt.plot(logits_range, softmax_outputs[:, 0], label="Class 1")
plt.plot(logits_range, softmax_outputs[:, 1], label="Class 2")
plt.plot(logits_range, softmax_outputs[:, 2], label="Class 3")
plt.title("Softmax Outputs for 3 Classes")
plt.xlabel("Logits")
plt.ylabel("Probability")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# Define the tanh function
def tanh(x):
    return np.tanh(x)

# Example input
x = np.array([-2, -1, 0, 1, 2])  # A range of inputs

# Apply the tanh function
tanh_output = tanh(x)

# Print the results
print("Input:", x)
print("Tanh Output:", tanh_output)
```

```python
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)  # Generate 100 points between -5 and 5

# Apply tanh function
y = tanh(x)

# Plot the tanh function
plt.plot(x, y, label="Tanh Function")
plt.title("Tanh Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Tanh Output (tanh(x))")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np

# Define the ReLU function
def relu(x):
    return np.maximum(0, x)

# Example input
x = np.array([-3, -1, 0, 1, 3])  # A range of inputs

# Apply the ReLU function
relu_output = relu(x)

# Print the results
print("Input:", x)
print("ReLU Output:", relu_output)
```

```python
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)  # Generate 100 points between -5 and 5

# Apply ReLU function
y = relu(x)

# Plot the ReLU function
plt.plot(x, y, label="ReLU Function")
plt.title("ReLU Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("ReLU Output (max(0, x))")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np

# Define the Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Example input
x = np.array([-3, -1, 0, 1, 3])  # A range of inputs

# Apply the Leaky ReLU function
alpha = 0.1  # Set alpha value (leakage slope)
leaky_relu_output = leaky_relu(x, alpha)

# Print the results
print("Input:", x)
print("Leaky ReLU Output:", leaky_relu_output)
```

```python
import matplotlib.pyplot as plt

# Generate input values
x = np.linspace(-5, 5, 100)  # Generate 100 points between -5 and 5

# Apply Leaky ReLU function
alpha = 0.1  # Set alpha value (leakage slope)
y = leaky_relu(x, alpha)

# Plot the Leaky ReLU function
plt.plot(x, y, label=f"Leaky ReLU (α={alpha})")
plt.title("Leaky ReLU Activation Function")
plt.xlabel("Input (x)")
plt.ylabel("Leaky ReLU Output")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid()
plt.legend()
plt.show()
```
