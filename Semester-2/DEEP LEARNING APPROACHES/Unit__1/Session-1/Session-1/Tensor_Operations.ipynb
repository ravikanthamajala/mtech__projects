```python
import tensorflow as tf

# Create a 1D tensor
tensor_1d = tf.constant([1, 2, 3, 4, 5])
print("1D Tensor:", tensor_1d)

# Create a 2D tensor
tensor_2d = tf.constant([[1, 2], [3, 4], [5, 6]])
print("\n2D Tensor:")
print(tensor_2d)

# Perform operations on tensors
# Add two tensors
tensor_sum = tf.add(tensor_2d, 2)
print("\nTensor after addition:")
print(tensor_sum)

# Multiply tensors element-wise
tensor_mul = tf.multiply(tensor_2d, 2)
print("\nTensor after element-wise multiplication:")
print(tensor_mul)

# Matrix multiplication
matrix1 = tf.constant([[1, 2], [3, 4]])
matrix2 = tf.constant([[5, 6], [7, 8]])
tensor_matmul = tf.matmul(matrix1, matrix2)
print("\nTensor after matrix multiplication:")
print(tensor_matmul)

# Get the shape and datatype of a tensor
print("\nShape of 2D Tensor:", tensor_2d.shape)
print("Data type of 2D Tensor:", tensor_2d.dtype)
```
