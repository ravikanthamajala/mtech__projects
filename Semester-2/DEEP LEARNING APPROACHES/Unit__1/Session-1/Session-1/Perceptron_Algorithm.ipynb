```python
import numpy as np

# Initialize variables
x = np.array([1, -1])  # Input features (x1, x2)
w = np.array([0.3, -0.1])  # Initial weights (w1, w2)
theta = 0.2  # Initial bias
alpha = 0.1  # Learning rate
Y_desired = 0  # Desired output

# Step function
def step_function(weighted_sum):
    return 1 if weighted_sum >= 0 else 0

# Perceptron training step-by-step
print("Initial weights:", w)
print("Initial bias:", theta)

# Compute the weighted sum
weighted_sum = np.dot(x, w) - theta
print("\nStep 1: Weighted sum =", weighted_sum)

# Apply the step activation function
Y_estimated = step_function(weighted_sum)
print("Step 2: Estimated output =", Y_estimated)

# Calculate the error
error = Y_desired - Y_estimated
print("Step 3: Error =", error)

# Update weights and bias if there is an error
if error != 0:
    delta_w = alpha * error * x
    w += delta_w
    delta_theta = -alpha * error
    theta += delta_theta

print("\nStep 4: Updated weights =", w)
print("Updated bias =", theta)
```
