```python
import tensorflow as tf

# Create a variable
var = tf.Variable([1, 2, 3, 4], dtype=tf.float32)
print("Variable:", var)

# Convert variable to constant
const = tf.convert_to_tensor(var)
print("Converted to constant:", const)
```

```python
# Create a constant
const = tf.constant([5, 6, 7, 8], dtype=tf.float32)
print("Constant:", const)

# Convert constant to variable
var = tf.Variable(const)
print("Converted to variable:", var)
```

```python
# Create a tensor
tensor = tf.constant([[1, 2, 3], [4, 5, 6]])
print("Original Tensor:")
print(tensor)

# Reshape to (3, 2)
reshaped_tensor = tf.reshape(tensor, [3, 2])
print("\nReshaped Tensor (3x2):")
print(reshaped_tensor)
```

```python
# Create a 2D tensor
tensor = tf.constant([[1, 2, 3], [4, 5, 6]])
print("Original Tensor:")
print(tensor)

# Flatten the tensor
flattened_tensor = tf.reshape(tensor, [-1])
print("\nFlattened Tensor:")
print(flattened_tensor)
```

```python
# Create a tensor
tensor = tf.constant([1, 2, 3, 4, 5, 6, 7, 8])

# Reshape to (2, 4)
reshaped_tensor1 = tf.reshape(tensor, [2, 4])
print("Reshaped Tensor (2x4):")
print(reshaped_tensor1)

# Reshape to (4, 2)
reshaped_tensor2 = tf.reshape(tensor, [4, 2])
print("\nReshaped Tensor (4x2):")
print(reshaped_tensor2)

# Reshape with -1 (let TensorFlow infer one dimension)
reshaped_tensor3 = tf.reshape(tensor, [-1, 2])  # Automatically infer rows
print("\nReshaped Tensor (-1, 2):")
print(reshaped_tensor3)
```
